```python
#acttivate geopandas_env

from pathlib import Path
import geopandas as gpd

in_dir = Path(r"C:\HA\CSA_N2O\Field_boundary\shp_to_geoparquet\shp_sk_test")
out_dir = in_dir.parent / "geoparquet_sk_test"
out_dir.mkdir(parents=True, exist_ok=True)

print("Input:", in_dir)
print("Output:", out_dir)
```

```python
shp_files = sorted(in_dir.glob("*.shp"))
print("Found shapefiles:", len(shp_files))

for shp in shp_files:
    try:
        gdf = gpd.read_file(shp)

        # Optional: if CRS is missing but you know it, uncomment and set it:
        # if gdf.crs is None:
        #     gdf = gdf.set_crs("EPSG:4326", allow_override=True)

        # Optional: fix invalid geometries (safe, but can be slower)
        # gdf["geometry"] = gdf["geometry"].make_valid()

        out_fp = out_dir / (shp.stem + ".geoparquet")
        gdf.to_parquet(out_fp, index=False)

        print(f"✅ {shp.name}  ->  {out_fp.name}   (rows={len(gdf)})")

    except Exception as e:
        print(f"❌ Failed: {shp.name} | {e}")
```

```python
# Load the first output file to confirm it's readable
out_files = sorted(out_dir.glob("*.geoparquet"))
print("GeoParquet files:", len(out_files))

test_fp = out_files[0]
gdf_check = gpd.read_parquet(test_fp)

print("Test file:", test_fp.name)
print("CRS:", gdf_check.crs)
gdf_check.head()
```
